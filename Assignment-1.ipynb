```python
import pandas as pd
import random
from faker import Faker

fake = Faker()

# Number of rows
num_rows = 1_000_000

departments = ['HR', 'IT', 'Finance', 'Marketing', 'Sales']

data = {
    "id": range(1, num_rows + 1),
    "name": [fake.name() for _ in range(num_rows)],
    "age": [random.randint(21, 60) for _ in range(num_rows)],
    "salary": [random.randint(30000, 150000) for _ in range(num_rows)],
    "department": [random.choice(departments) for _ in range(num_rows)]
}

df = pd.DataFrame(data)

# Save to CSV
df.to_csv("employees.csv", index=False)

print("CSV file with 1 million rows created successfully!")
```

```python
import dask.dataframe as dd

# Load CSV into Dask DataFrame
dask_df = dd.read_csv("employees.csv")

print("Dask DataFrame Loaded Successfully!")
```
